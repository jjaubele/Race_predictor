```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_t800(t400, t1500):
    return t400 * 1.2 + t1500 * 0.216

def calculate_p800(p400, p1500):
    return p400 * 0.597 + p1500 * 0.403
```

```python
# Graficar t1500 en funcin de t400 para un t800 fijo
t800_fixed = 120  # segundos
t400_range = np.linspace(48, 60, 100)  # t400 entre 48s y 60s

# Despejar t1500 de la ecuacin: t800 = t400 * 1.2 + t1500 * 0.216
# t1500 = (t800 - t400 * 1.2) / 0.216
t1500_values = (t800_fixed - t400_range * 1.2) / 0.216

plt.figure(figsize=(10, 6))
plt.plot(t400_range, t1500_values, linewidth=2)
plt.xlabel('Tiempo 400m (s)', fontsize=12)
plt.ylabel('Tiempo 1500m (s)', fontsize=12)
plt.title(f'Relacin t400 vs t1500 para t800 = {t800_fixed}s', fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()
```
